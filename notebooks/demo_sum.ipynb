# Demo Notebook: Sum of 5 Numbers

This notebook demonstrates a simple calculation of summing 5 numbers.

```python
# Define 5 numbers
numbers = [10, 25, 30, 45, 50]

print("Numbers:", numbers)
```

```python
# Calculate the sum
total = sum(numbers)

print(f"Sum of {numbers} = {total}")
```

```python
# Alternative method using a loop
total_loop = 0
for num in numbers:
    total_loop += num

print(f"Sum using loop: {total_loop}")
```

```python
# Using numpy
import numpy as np

numbers_array = np.array(numbers)
total_numpy = np.sum(numbers_array)

print(f"Sum using NumPy: {total_numpy}")
```

## Summary

All three methods produce the same result:
- Built-in `sum()` function
- Manual loop iteration
- NumPy's `np.sum()` function
